```python
import networkx as nx
import matplotlib.pyplot as plt
import random

def generate_graph(n, p=0.3):
    G = nx.erdos_renyi_graph(n, p)
    return G

def assign_trd3_labels(G):
    labels = {}
    for node in G.nodes():
        labels[node] = random.choice([0, 1, 2])  # Assign TRD3 values
    return labels

def evaluate_trd3(G, labels):
    domination_value = 0
    for node in G.nodes():
        if labels[node] == 2:
            domination_value += 2
        elif labels[node] == 1:
            if any(labels[neighbor] > 0 for neighbor in G.neighbors(node)):
                domination_value += 1
    return domination_value

def draw_graph(G, labels):
    pos = nx.spring_layout(G)
    node_colors = ['red' if labels[node] == 2 else 'blue' if labels[node] == 1 else 'gray' for node in G.nodes()]

    nx.draw(G, pos, with_labels=True, node_color=node_colors, edge_color='black', node_size=800, font_size=10)

    node_labels = {node: str(labels[node]) for node in G.nodes()}
    nx.draw_networkx_labels(G, pos, labels=node_labels, font_color='white')

    plt.title("Total Roman Domination 3 Graph")
    plt.show()

# Generate graph
graph = generate_graph(10, 0.3)
labels = assign_trd3_labels(graph)

domination_value = evaluate_trd3(graph, labels)
print(f"Total Roman Domination Value: {domination_value}")

draw_graph(graph, labels)
#red-->> 2 stronest
#blue-->> 1 moderate
#grey-->> 0
```
